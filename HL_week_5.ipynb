```python
import altair as alt
import pandas as pd
```

```python
math = pd.read_csv("prueba6.csv",  sep=";")
print(math)
```

```python
brush = alt.selection_interval()
```

```python
points = alt.Chart(math).mark_point().encode(
    x='G3:Q',
    y='absences:Q',
    color=alt.condition(brush, 'sex:N', alt.value('lightgray'))
).add_selection(
    brush
)

points
```

```python
bars1 = alt.Chart(math).mark_bar().encode(
    y='sex:N',
    color='sex:N',
    x='count(sex):Q'
)

bars1

#see how many men and women are studying
```
